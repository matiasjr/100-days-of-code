```python
import numpy as np

# Extrair as coordenadas dos centróides no formato correto (latitude, longitude)
coord_am = [(point.y, point.x) for point in centroids]  # Latitudes e longitudes

# Usar a classe KNN do libpysal para calcular os k-vizinhos mais próximos
k1_weights = KNN.from_dataframe(am_muni, k=1, geom_col="geometry")

#-- Passo 4: Calcular as distâncias entre os vizinhos --#
# Função para calcular a distância do Grande Círculo entre dois pontos
def calculate_great_circle_distance(coord1, coord2):
    return great_circle(coord1, coord2).kilometers

# Extrair as distâncias entre os vizinhos
k1_dists = []
for i, neighbors in k1_weights.neighbors.items():
    for neighbor in neighbors:
        dist = calculate_great_circle_distance(coord_am[i], coord_am[neighbor])
        k1_dists.append(dist)

# Resumo das distâncias
k1_dists = np.array(k1_dists)
print("Resumo das distâncias:")
print(f"Min.: {k1_dists.min():.2f}")
print(f"1st Qu.: {np.percentile(k1_dists, 25):.2f}")
print(f"Median: {np.median(k1_dists):.2f}")
print(f"Mean: {k1_dists.mean():.2f}")
print(f"3rd Qu.: {np.percentile(k1_dists, 75):.2f}")
print(f"Max.: {k1_dists.max():.2f}")

#-- Passo 5: Calcular os vizinhos dentro de um limite de distância (0 a 233 km) --#
# Usar a classe DistanceBand do libpysal para calcular os vizinhos dentro de 233 km
threshold_distance_km = 233

# Criar uma lista de vizinhos baseada na distância do Grande Círculo
neighbors_dict = {}
for i, coord_i in enumerate(coord_am):
    neighbors = []
    for j, coord_j in enumerate(coord_am):
        if i != j:  # Ignorar o próprio município
            dist = calculate_great_circle_distance(coord_i, coord_j)
            if dist <= threshold_distance_km:
                neighbors.append(j)
    neighbors_dict[i] = neighbors


#-- Passo 4: Verificar a conectividade da matriz de pesos --#
# Contar componentes desconectados
connected_components = []
visited = set()

def dfs(node, neighbors_dict, visited, component):
    visited.add(node)
    component.append(node)
    for neighbor in neighbors_dict[node]:
        if neighbor not in visited:
            dfs(neighbor, neighbors_dict, visited, component)

for node in range(len(coord_am)):
    if node not in visited:
        component = []
        dfs(node, neighbors_dict, visited, component)
        connected_components.append(component)

print(f"Número de componentes desconectados: {len(connected_components)}")

# Exibir a estrutura da lista de vizinhos
print(f"\nLista de vizinhos dentro de {threshold_distance_km} km:")
print(f"Número de componentes desconectados: {len(connected_components)}")
print(k1_dists)
```

```python
from geopy.distance import great_circle

# Calcular as distâncias entre todos os pares de municípios
distances = []
for i in range(len(coord_am)):
    row = []
    for j in range(len(coord_am)):
        if i == j:
            row.append(0)  # Distância para si mesmo é zero
        else:
            dist = calculate_great_circle_distance(coord_am[i], coord_am[j])
            row.append(dist)
    distances.append(row)

print(distances)
```

```python
# Criar uma matriz de pesos baseada no inverso da distância
weights = []
for i in range(len(distances)):
    row = []
    for j in range(len(distances[i])):
        if i == j:
            row.append(0)  # Peso para si mesmo é zero
        elif distances[i][j] > 233:  # Ignorar distâncias maiores que 233 km
            row.append(0)
        else:
            row.append(1 / distances[i][j])  # Inverso da distância
    weights.append(row)

# Converter para um array NumPy para facilitar manipulação
weights = np.array(weights)
```

```python
# Normalizar os pesos
row_sums = weights.sum(axis=1)
normalized_weights = weights / row_sums[:, np.newaxis]
```

```python
from libpysal.weights import W
import networkx as nx

# Criar um objeto W com os pesos normalizados
# Garantir que 'weights' seja um dicionário alinhado com 'neighbors_dict'
weights_dict = {}
for i, neighbors in neighbors_dict.items():
    # Extrair os pesos correspondentes aos vizinhos
    weights_dict[i] = [normalized_weights[i][j] for j in neighbors]

# Criar um objeto W com os pesos normalizados
w = W(neighbors_dict, weights_dict)

# Converter para um grafo NetworkX
G = w.to_networkx()

# Convert to an undirected graph
G = nx.Graph(G) # Convert to undirected graph

# Identificar componentes conectados
connected_components = list(nx.connected_components(G))
print(f"Número de componentes desconectados: {len(connected_components)}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from libpysal.weights import Queen, higher_order, W
from esda.moran import Moran

# Criar a matriz de vizinhança baseada na regra da rainha
lvizQ = Queen.from_dataframe(am_muni, ids='code_muni')

# Definir o número de lags espaciais
max_lag = 6

# Função para calcular o correlograma espacial
def calculate_spatial_correlogram(weights, values, max_lag, permutations=999):
    moran_values = []
    p_values = []

    for lag in range(1, max_lag + 1):
        # Criar uma matriz de pesos para o lag atual (vizinhos de ordem superior)
        lag_W = higher_order(weights, lag)

        # Calcular Moran's I para o lag atual
        moran = Moran(values, lag_W, permutations=permutations)
        moran_values.append(moran.I)
        p_values.append(moran.p_sim)

    return moran_values, p_values

# Calcular o correlograma para a matriz padronizada (W)
moran_W, p_W = calculate_spatial_correlogram(lvizQ, media_ideb, max_lag, permutations=999)

# Transformar a matriz em binária (B)
lvizQ.transform = 'B'
moran_B, p_B = calculate_spatial_correlogram(lvizQ, media_ideb, max_lag, permutations=999)

# Plotar os gráficos
lags = np.arange(1, max_lag + 1)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico para matriz padronizada (W)
axes[0].errorbar(lags, moran_W, yerr=p_W, fmt='o', capsize=5, label="Moran's I")
axes[0].axhline(y=0, color='gray', linestyle='--')
axes[0].set_title("Ideb Amazonas 2021, Proximidade Padronizada")
axes[0].set_xlabel("Lags")
axes[0].set_ylabel("Moran's I")
axes[0].legend()

# Gráfico para matriz binária (B)
axes[1].errorbar(lags, moran_B, yerr=p_B, fmt='o', capsize=5, label="Moran's I")
axes[1].axhline(y=0, color='gray', linestyle='--')
axes[1].set_title("Ideb Amazonas 2021, Proximidade Binária")
axes[1].set_xlabel("Lags")
axes[1].set_ylabel("Moran's I")
axes[1].legend()

# Ajustar layout
plt.tight_layout()
plt.show()
```
